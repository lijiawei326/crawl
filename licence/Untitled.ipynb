```python
import pandas as pd
```

```python
df = pd.read_csv('./guangxi_licence.csv')
df
```

```python
def add(x):
    if len(x) == 0:
        return x
    else:
        y = []
        for i in x:
            if str(i) == 'nan':
                continue
            y.append(str(i))
        return '、'.join(y)
a = df.groupby(df.columns[:13].to_list()).agg(add)
```

```python
a = a.reset_index()
```

```python
a.to_csv('./zhejiang_licence1.csv',index=False)
```

```python
for i in range(len(a)):
    a.loc[i,'品种明细'] = a.loc[i,'品种明细'].replace('8','(').replace('\x029',')')
```
